```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import cvxopt
from cvxopt import matrix
from cvxopt import solvers
```

**<font color = blue size=4>Á¨¨‰∏ÄÈÉ®ÂàÜ:ÂáΩÊï∞‰ªãÁªç</font>**

‰∫åÊ¨°ËßÑÂàíÈóÆÈ¢òÊòØÂΩ¢ÂºèÂ¶Ç‰∏ãÁöÑ‰∏ÄÁ±ªÊúÄ‰ºòÂåñÈóÆÈ¢òÔºö
$$
\begin{align}
\min_x \quad  &\frac{1}{2}x^TPx+q^Tx \\
s.t. \quad  &Gx\leq h \\
      &Ax=b
\end{align}
$$
ÂØπ‰∫éËøô‰∏ÄÁ±ªÈóÆÈ¢òÂèØ‰ª•‰ΩøÁî®[cvxopt](https://cvxopt.org/userguide/coneprog.html#quadratic-programming)Â∫ìÁöÑsolvers.qp()ÂáΩÊï∞ËøõË°åÊ±ÇËß£„ÄÇ

‰ª•‰∏ãÊòØ‰∏Ä‰∏™‰æãÂ≠êÔºàÂèÇËÄÉ[Solving a quadratic program](https://cvxopt.org/examples/tutorial/qp.html)Ôºâ:
$$
\begin{align}
\min_x \quad  &2x_1^2+x_2^2+x_1x_2+x_1+x_2 \\
s.t. \quad  &x_1\geq 0 \\
      &x_2\geq 0 \\
      &x_1+x_2=1
\end{align}
$$
‰∏∫‰∫Ü‰ΩøÁî®solvers.qp()ÂáΩÊï∞ÔºåÊàë‰ª¨ÈúÄË¶ÅÁü•ÈÅìÂú®ËØ•‰∫åÊ¨°ËßÑÂàíÈóÆÈ¢ò‰∏≠ÁöÑ$P,q,G,h,A,b$Áü©ÈòµÂàÜÂà´ÊòØ‰ªÄ‰πà„ÄÇ
Âú®ËØ•‰ºòÂåñÈóÆÈ¢ò‰∏≠Ôºå

* $P:=\begin{bmatrix}
    4 & 1 \\ 1 & 2
   \end{bmatrix}$,
* $q:=\begin{bmatrix}
    1 \\ 1
   \end{bmatrix}$,
* $G:=\begin{bmatrix}
    -1 & 0 \\ 0 & -1
   \end{bmatrix}$,
* $h:=\begin{bmatrix}
    0 \\ 0
   \end{bmatrix}$,
* $A:=\begin{bmatrix}
    1 & 1
   \end{bmatrix}$,
* $b:=\begin{bmatrix}
    1
   \end{bmatrix}$,
   
ÊääËøô‰∫õÂèÇÊï∞ÈÄÅÂÖ•solvers.qp()ÂáΩÊï∞‰∏≠Âç≥ÂèØÊ±ÇÂá∫Ëß£„ÄÇ

```python
# Tips1: cvxoptÂ∫ì‰∏≠ÁöÑmatrixÂè™Êé•ÂèódoubleÁ±ªÂûãÁöÑÊï∞ÊçÆ
# Tips2: matrix‰ΩøÁî®ÂàóË°®‰Ωú‰∏∫ÂèÇÊï∞ÂàõÂª∫Áü©ÈòµÂíånumpy.array‰ΩøÁî®ÂàóË°®‰Ωú‰∏∫ÂèÇÊï∞ÂàõÂª∫Áü©ÈòµÊòØ‰∏çÂêåÁöÑ
# print(matrix([[1.0, 1.0]]))
# print(np.array([[1.0, 1.0]]))
# print(matrix(np.array([[1.0, 1.0]])))
P = 2*matrix([ [2, .5], [.5, 1] ])
q = matrix([1.0,1.0])
G = matrix([[-1.0,0.0],[0.0,-1.0]])
h = matrix([0.0,0.0])
A = matrix([1.0, 1.0], (1,2))
b = matrix(1.0)
sol=solvers.qp(P, q, G, h, A, b)
print(sol['x'])
```

1.Á∫øÊÄßÂèØÂàÜÊîØÊåÅÂêëÈáèÊú∫‰∏éÁ°¨Èó¥ÈöîÊúÄÂ§ßÂåñ

<span style="color:purple">1) ËØªÂÖ•Êï∞ÊçÆÈõÜ'dataset1.csv',ÊääÊï∞ÊçÆÁ±ªÂûãÈÉΩËΩ¨Êç¢Êàênp.doubleÁ±ªÂûãÔºåÂπ∂ÁîªÂá∫Êï∞ÊçÆÈõÜÁöÑÊï£ÁÇπÂõæÔºåÁªôÊ≠£Ê†∑Êú¨Ôºày‰∏∫+1ÔºâÂíåË¥üÊ†∑Êú¨Ôºày‰∏∫-1ÔºâÂàÜÂà´Ê†á‰∏ä‰∏çÂêåÁöÑÈ¢úËâ≤„ÄÇ</span>

```python
# ---- Your code here ----
data1 = pd.read_csv("dataset1.csv",dtype = np.double)
data_y_1 = data1[data1['y']==1.0]
data_y_2 = data1[data1['y']==-1.0]
x1_1 = data_y_1["x1"]
x2_1 = data_y_1["x2"]
plt.scatter(x1_1,x2_1,c="r")
x1_2 = data_y_2["x1"]
x2_2 = data_y_2["x2"]
plt.scatter(x1_2,x2_2,c="b")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

<span style="color:purple">2) Ê±ÇËß£Â¶Ç‰∏ãÂØπÂÅ∂ÈóÆÈ¢òÔºàÂèÇËÄÉËØæ‰ª∂ÔºâÔºö</span>
$$
\begin{align}
\min_\alpha \quad  &\frac{1}{2}\sum_{i=1}^m\sum_{j=1}^m\alpha_i\alpha_jy_iy_j\pmb{x}_i^T\pmb{x}_j-\sum_{i=1}^m\alpha_i \\
s.t. \quad  &\sum_{i=1}^m\alpha_iy_i=0 \\
      &\pmb{\alpha}\geq \pmb{0}
\end{align}
$$

Ëøô‰∏™‰ºòÂåñÈóÆÈ¢òÊòØ‰∏Ä‰∏™‰∫åÊ¨°ËßÑÂàíÈóÆÈ¢ò„ÄÇ
* $PÊòØ‰∏Ä‰∏™m\times mÁöÑÁü©ÈòµÔºåÂÖ∂‰∏≠P_{ij}=y_iy_j\pmb{x}_i^T\pmb{x}_j$,
* $qÊòØ‰∏Ä‰∏™m\times 1ÁöÑÊâÄÊúâÂÄºÈÉΩ‰∏∫-1ÁöÑÂàóÂêëÈáèÔºåÂç≥q:=\begin{bmatrix}
    -1 & -1 & \cdots & -1
   \end{bmatrix}^T$,
* $G:=\begin{bmatrix}
    -1 & 0 & \cdots & 0 \\
    0 & -1 & \cdots & 0 \\
   \vdots & \vdots & \ddots &0 \\
   0 & 0 & 0 & -1
   \end{bmatrix}_{m\times m}=-\pmb{I},\pmb{I}‰∏∫Âçï‰ΩçÁü©Èòµ,$
* $hÊòØ‰∏Ä‰∏™m\times 1ÁöÑÈõ∂ÂêëÈáè,Âç≥h:=\begin{bmatrix}
    0 & 0 & \cdots & 0
   \end{bmatrix}^T$,
* $A:=\begin{bmatrix}
    y_1 & y_2 & \cdots & y_m
   \end{bmatrix}^T$,
* $b:=\begin{bmatrix}
    0
   \end{bmatrix},‰∏Ä‰∏™Ê†áÈáè$
   
<span style="color:purple">Êää‰∏äËø∞ÂèÇÊï∞ÈÄÅÂÖ•Ê±ÇËß£Âô®solvers.qp()‰∏≠Âç≥ÂèØÂæóÂà∞ÊúÄ‰ºòËß£$\alpha^*$</span>„ÄÇ 
 
ÈôÑÔºö$P$Áü©ÈòµÁöÑ‰∏Ä‰∏™ËÆ°ÁÆóÊñπÊ≥ïÔºö
ËÆæ$X=\begin{bmatrix}
    x_{11} & x_{12} \\
    x_{21} & x_{22} \\
    \vdots & \vdots \\
    x_{m1} & x_{m2}
   \end{bmatrix}$,
   $Y=\begin{bmatrix}
    y_{1} \\
    y_{2} \\
    \vdots \\
    y_{m}
   \end{bmatrix}$,
   
ËÆ°ÁÆó$X'=\begin{bmatrix}
    x_{11}y_1 & x_{12}y_1 \\
    x_{21}y_2 & x_{22}y_2 \\
    \vdots & \vdots \\
    x_{m1}y_m & x_{m2}y_m
   \end{bmatrix}=X*Y(Ê≥®ÊÑèËøôÈáåÊòØÊòü‰πò)$
   
Âàô$P=X'X'^T$„ÄÇ

```python
# ---- Your code here ----
#Â¶ÇÊûúÊ±ÇËß£Êä•ÈîôÂèØ‰ª•Â∞ùËØïÂú®solvers.qp()‰∏≠Ê∑ªÂä†ÂèÇÊï∞kktsolver='ldl'
m = 40        # data1.info() Ëé∑Âæó‰∏ÄÂÖ±Êúâ40‰∏™Ê†∑Êú¨
X = np.matrix(data1[["x1","x2"]])
Y = np.matrix(data1[["y"]])
X_ = np.multiply(X,Y)
P = matrix(np.dot(X_,X_.T))
q = matrix((-1)*np.ones((40,1)))
G = matrix((-1)*np.eye(40,40))
h = matrix(np.zeros((40,1)))
A = matrix(Y.T)
b = matrix(0.0)
sol=solvers.qp(P, q, G, h, A, b,kktsolver='ldl')
a = sol["x"]
print(sol["x"])
```

<span style="color:purple">3) Ê±ÇÂá∫$\pmb{\omega}^*=\sum_{i=1}^m\alpha_i^*y_i\pmb{x}_i$Âíå$b^*=y_j-\pmb{\omega}^{*T}\pmb{x_j}$, ÂÖ∂‰∏≠$j$‰∏∫$\alpha^*$‰∏≠ÁöÑ‰∏Ä‰∏™Ê≠£ÂàÜÈáè$\alpha_j^*>0$ÁöÑ‰∏ãÊ†á„ÄÇÊ≥®ÊÑèÔºöÁî±‰∫éÊ±ÇËß£Âô®Ê±ÇÂá∫Êù•ÁöÑÊòØ‰∏Ä‰∏™Ëøë‰ººËß£ÔºåÊâÄ‰ª•$\alpha^*$‰∏≠ÂæàÂ§öÂÆûÈôÖ‰∏ä‰∏∫0ÁöÑÂàÜÈáè‰ºöÁï•Â§ß‰∫é0ÔºåËøôÊó∂ÂÄôÂèØ‰ª•ËÆæÁΩÆ‰∏Ä‰∏™ÈòàÂÄºÊääÈùûÂ∏∏Èù†Ëøë0ÁöÑÈÇ£‰∫õÂàÜÈáèÁ≠õÂéªÔºåÂÜç‰ªéÂâ©‰∏ãÁöÑÂàÜÈáè‰∏≠ÈÄâÂèñ‰∏Ä‰∏™Ê≠£ÂàÜÈáèÊù•ËÆ°ÁÆó$b^*$,ÊàñËÄÖ‰πüÂèØ‰ª•Áõ¥Êé•Âèñ$\alpha^*$‰∏≠ÊúÄÂ§ßÁöÑÂàÜÈáèÊù•ËÆ°ÁÆó$b^*$„ÄÇ</span>

```python
matrix_1 = np.multiply(a,Y)
w = np.multiply(matrix_1,X)
w1_best = w[:,0].sum()
w2_best = w[:,1].sum()
w_best = matrix([w1_best,w2_best])
a_list = list(a)
index_max = a_list.index(np.max(a_list))
b_best = matrix(matrix(Y)[index_max] - w_best.T * X[index_max].T)
b_best = b_best[0]
print(w_best)
print(b_best)
```

```python
#ËÆæÁΩÆÈòàÂÄº Êù•‰ΩøÂæóÂæàÂ∞èÁöÑùõº‰∏∫0.
min_data = 1e-8
l = list()                      # list Êù•‰øùÂ≠òùõº‰∏ç‰∏∫0ÁöÑÁ¥¢Âºï
for i in range(len(a)):
    if a[i]<min_data:
        a[i]=0
    else:
        l.append(i)
print(a)
print(l)
```

<span style="color:purple">4) ÁîªÂá∫Êï∞ÊçÆÈõÜÁöÑÊï£ÁÇπÂõæÔºåÁªôÊ≠£Ê†∑Êú¨Ôºày‰∏∫+1ÔºâÂíåË¥üÊ†∑Êú¨Ôºày‰∏∫-1ÔºâÂàÜÂà´Ê†á‰∏ä‰∏çÂêåÁöÑÈ¢úËâ≤ÔºåÂÜç‰∏∫ÊîØÊåÅÂêëÈáèÔºàËÆ≠ÁªÉÊï∞ÊçÆ‰∏≠$\alpha_j^*>0$ÁöÑÂØπÂ∫îÁöÑÊ†∑Êú¨ÔºâÊ†á‰∏ä‰∏çÂêåÁöÑÈ¢úËâ≤ÔºåÂπ∂ÁîªÂá∫ÂÜ≥Á≠ñËæπÁïå$\pmb{\omega}^{*T}\pmb{x}+b=0$ÂíåÈó¥ÈöîËæπÁïå$\pmb{\omega}^{*T}\pmb{x}+b=1$‰∏é$\pmb{\omega}^{*T}\pmb{x}+b=-1$„ÄÇ</span>

```python
# ---- Your code here ----
# ÊâæÂá∫ÊîØÊåÅÂêëÈáè
plt.figure(figsize=(6, 6))
l1 = data_y_1.index.tolist()
l2 = data_y_2.index.tolist()
l2.sort(reverse=True)
print(l2)

data_y_1 = data1[data1['y']==1.0]
data_y_2 = data1[data1['y']==-1.0]
x1_1 = data_y_1["x1"]
x2_1 = data_y_1["x2"]
c1 = ["purple" if i in l else "red" for i in l1] 
plt.scatter(x1_1,x2_1,c=c1,label="+")
x1_2 = data_y_2["x1"].sort_index(ascending = False)
x2_2 = data_y_2["x2"].sort_index(ascending = False)
c2 = ["purple" if j in l else "blue" for j in l2]
plt.scatter(x1_2,x2_2,c=c2,label = "-")
x_list = np.linspace(-4,8)
y1_list = -x_list *w1_best/w2_best-b_best/w2_best
plt.plot(x_list,y1_list,c ="orange",label="WX+b=0")
y2_list = -x_list *w1_best/w2_best-(b_best-1)/w2_best
plt.plot(x_list,y2_list,c ="black",label="WX+b=1")
y3_list = -x_list *w1_best/w2_best-(b_best+1)/w2_best
plt.plot(x_list,y3_list,c ="green",label="WX+b=-1")
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()
plt.show()
```

<span style="color:purple">1) ËØªÂÖ•Êï∞ÊçÆÈõÜ'dataset2.csv',ÊääÊï∞ÊçÆÁ±ªÂûãÈÉΩËΩ¨Êç¢Êàênp.doubleÁ±ªÂûãÔºåÂπ∂ÁîªÂá∫Êï∞ÊçÆÈõÜÁöÑÊï£ÁÇπÂõæÔºåÁªôÊ≠£Ê†∑Êú¨Ôºày‰∏∫+1ÔºâÂíåË¥üÊ†∑Êú¨Ôºày‰∏∫-1ÔºâÂàÜÂà´Ê†á‰∏ä‰∏çÂêåÁöÑÈ¢úËâ≤„ÄÇ</span>

```python
# ---- Your code here ----
data2 = pd.read_csv("dataset2.csv",dtype=np.double)
data_y_1 = data2[data2['y']==1.0]
data_y_2 = data2[data2['y']==-1.0]
x1_1 = data_y_1["x1"]
x2_1 = data_y_1["x2"]
plt.scatter(x1_1,x2_1,c="r")
x1_2 = data_y_2["x1"]
x2_2 = data_y_2["x2"]
plt.scatter(x1_2,x2_2,c="b")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

<span style="color:purple">2) ÈÄâÊã©‰∏Ä‰∏™ÂèÇÊï∞CÔºåÊ±ÇËß£Â¶Ç‰∏ãÂØπÂÅ∂ÈóÆÈ¢òÔºàÂèÇËÄÉËØæ‰ª∂ÔºâÔºö</span>
$$
\begin{align}
\min_\alpha \quad  &\frac{1}{2}\sum_{i=1}^m\sum_{j=1}^m\alpha_i\alpha_jy_iy_j\pmb{x}_i^T\pmb{x}_j-\sum_{i=1}^m\alpha_i \\
s.t. \quad  &\sum_{i=1}^m\alpha_iy_i=0 \\
      &\pmb{0}\leq \pmb{\alpha}\leq C 
\end{align}
$$
* $G:=\begin{bmatrix}
    -1 & 0 & \cdots & 0 \\
    0 & -1 & \cdots & 0 \\
   \vdots & \vdots & \ddots &0 \\
   0 & 0 & 0 & -1 \\
   1 & 0 & \cdots & 0 \\
    0 & 1 & \cdots & 0 \\
   \vdots & \vdots & \ddots &0 \\
   0 & 0 & 0 & 1
   \end{bmatrix}_{2m\times m}=\begin{bmatrix}
    -\pmb{I} \\
    \pmb{I}
   \end{bmatrix},\pmb{I}‰∏∫Âçï‰ΩçÁü©Èòµ,$
* $h:=\begin{bmatrix}
    0 \\
    0 \\
    \vdots \\
    0 \\
    C \\
    C \\
    \vdots \\
    C
   \end{bmatrix}_{2m\times 1}, Âç≥‰∏Ä‰∏™m\times 1ÁöÑÈõ∂ÂàóÂêëÈáè‰∏é‰∏Ä‰∏™m\times 1ÁöÑÂàÜÈáèÂÖ®‰∏∫CÁöÑÂàóÂêëÈáè‰∏ä‰∏ãÊãºÊé•$,
* $P,q,A,b$‰∏éÁ°¨Èó¥Èöî‰ºòÂåñÈóÆÈ¢ò‰∏≠ÁöÑÁü©ÈòµÁõ∏Âêå„ÄÇ

```python
data2.info()                     # Ëé∑Âèñm
```

```python
# ---- Your code here ----
m = 40                             # data1.info() Ëé∑Âæó‰∏ÄÂÖ±Êúâ40‰∏™Ê†∑Êú¨
c = 0.5
X = np.matrix(data2[["x1","x2"]])
Y = np.matrix(data2[["y"]])
X_ = np.multiply(X,Y)
P = matrix(np.dot(X_,X_.T))
q = matrix((-1)*np.ones((40,1)))
G1 = matrix((-1)*np.eye(40,40))
G = matrix(np.vstack((G1,-G1)))
h1 = matrix(np.zeros((40,1)))
h2 = matrix(np.ones((40,1)))*c
h = matrix(np.vstack((h1, h2)))
A = matrix(Y.T)
b = matrix(0.0)
sol=solvers.qp(P, q, G, h, A, b,kktsolver='ldl')
a = sol["x"]
print(sol["x"])

```

<span style="color:purple">3) Ê±ÇÂá∫$\pmb{\omega}^*=\sum_{i=1}^m\alpha_i^*y_i\pmb{x}_i$Âíå$b^*=y_j-\pmb{\omega}^{*T}\pmb{x_j}$, ÂÖ∂‰∏≠$j$‰∏∫$\alpha^*$‰∏≠ÁöÑ‰∏Ä‰∏™Ê≠£ÂàÜÈáè$0<\alpha_j^*<C$ÁöÑ‰∏ãÊ†á„ÄÇ‰∏éÁ°¨Èó¥Èöî‰ºòÂåñÈóÆÈ¢òÂêåÁêÜÔºåÂ∫îËØ•Á≠õÊéâÈùûÂ∏∏Êé•Ëøë0ÂíåÈùûÂ∏∏Êé•Ëøë$C$ÁöÑÂàÜÈáè„ÄÇ</span>

```python
# ---- Your code here ----
matrix_1 = np.multiply(a,Y)
w = np.multiply(matrix_1,X)
w1_best = w[:,0].sum()
w2_best = w[:,1].sum()
w_best = matrix([w1_best,w2_best])
a_list = list(a)
flag_list = []
for i in a:
    if i>1e-8 and (0.5-i)>1e-8:
        flag_list.append(i)
index_max = a_list.index(np.max(flag_list))    
b_best = matrix(matrix(Y)[index_max] - w_best.T * X[index_max].T)
b_best = b_best[0]
print(w_best)
print(b_best)     



```

```python
#ËÆæÁΩÆÈòàÂÄº Êù•‰ΩøÂæóÂæàÂ∞èÁöÑùõº‰∏∫0.
min_data = 1e-7
l = list() # list Êù•‰øùÂ≠òùõº‰∏ç‰∏∫0ÁöÑÁ¥¢Âºï
l_ = list()
for i in range(len(a)):
    if a[i]<min_data:
        a[i]=0
    if a[i]>min_data:
        l.append(i)
        l_.append(a[i])
print(a)
print(l)
print(l_)
```

3.ÈùûÁ∫øÊÄßÊîØÊåÅÂêëÈáèÊú∫‰∏éÊ†∏ÂáΩÊï∞

[Raisin Dataset](https://www.kaggle.com/datasets/muratkokludataset/raisin-dataset)ÊòØ‰∏Ä‰∏™Ëë°ËêÑÂπ≤ÁöÑÊï∞ÊçÆÈõÜÔºåÊÄªÂÖ±Êúâ900‰∏™Ê†∑Êú¨ÔºåÊØè‰∏™Ê†∑Êú¨ÂåÖÂê´7‰∏™(ÈÉΩÊòØËøûÁª≠ÁöÑ)ÁâπÂæÅ‰ª•Âèä1‰∏™Ê†áÁ≠æÔºåÊØè‰∏™Ê†áÁ≠æÂè™Êúâ‰∏§ÁßçÂèØËÉΩÂèñÂÄº„ÄÇÊú¨Ê¨°ÂÆûÈ™åÂ∑≤ÁªèÊåâÁÖß8Ôºö2ÁöÑÊØî‰æãÂàíÂàÜÊàê‰∫ÜËÆ≠ÁªÉÊï∞ÊçÆÈõÜ'Raisin_train.csv'‰ª•ÂèäÊµãËØïÊï∞ÊçÆÈõÜ'Raisin_test.csv'Ôºå‰∏îÊØè‰∏™Êï∞ÊçÆÈõÜÈÉΩÂ∑≤ÁªèÂÅö‰∫ÜÁâπÂæÅÂΩí‰∏ÄÂåñÂ§ÑÁêÜ‰ª•ÂèäÊääÊ†áÁ≠æÁöÑÂÄºÊõøÊç¢Êàê‰∫Ü+1Âíå-1„ÄÇ

<span style="color:purple">1) ËØªÂÖ•ËÆ≠ÁªÉÊï∞ÊçÆÈõÜ'Raisin_train.csv',ÊääÊï∞ÊçÆÁ±ªÂûãÈÉΩËΩ¨Êç¢Êàênp.doubleÁ±ªÂûã„ÄÇ</span>

```python
# ---- Your code here ----
train_data = pd.read_csv("Raisin_train.csv")
test_data = pd.read_csv("Raisin_test.csv")
train_data.head()


```

<span style="color:purple">2) ÈÄâÊã©‰∏Ä‰∏™Ê†∏ÂáΩÊï∞$K(\pmb{x},\pmb{z})$‰ª•ÂèäÂèÇÊï∞CÔºåÊ±ÇËß£Â¶Ç‰∏ãÂØπÂÅ∂ÈóÆÈ¢òÔºàÂèÇËÄÉËØæ‰ª∂ÔºâÔºö</span>
$$
\begin{align}
\min_\alpha\quad   &\frac{1}{2}\sum_{i=1}^m\sum_{j=1}^m\alpha_i\alpha_jy_iy_jK(\pmb{x}_i,\pmb{x}_j)-\sum_{i=1}^m\alpha_i \\
s.t. \quad  &\sum_{i=1}^m\alpha_iy_i=0 \\
      &\pmb{0}\leq \pmb{\alpha}\leq C 
\end{align}
$$

Áõ∏ËæÉ‰∫éËΩØÈó¥ÈöîÊúÄÂ§ßÂåñÁöÑ‰ºòÂåñÈóÆÈ¢òÔºåËØ•‰ºòÂåñÈóÆÈ¢ò‰ªÖÈúÄË¶ÅÂØπÁü©Èòµ$P$ÂÅöÊîπÂä®„ÄÇ
‰ªé‰ª•‰∏ãÂ∏∏Áî®ÁöÑÊ†∏ÂáΩÊï∞‰∏≠ÈÄâÊã©‰∏Ä‰∏™‰Ωú‰∏∫ËØ•‰ºòÂåñÈóÆÈ¢ò‰∏≠ÁöÑ$K$ÔºàÂèÇÊï∞Ëá™Â∑±ËøõË°åË∞ÉÊï¥ÔºâÔºö
* Á∫øÊÄßÊ†∏Ôºö$K(\pmb{x},\pmb{z})=\pmb{x}^T\pmb{z}$
* Â§öÈ°πÂºèÊ†∏Ôºö$K(\pmb{x},\pmb{z})=(\pmb{x}^T\pmb{z}+1)^p$
* È´òÊñØÊ†∏Ôºö$K(\pmb{x},\pmb{z})=exp(-\frac{\parallel \pmb{x}-\pmb{z} \parallel^2}{2\sigma^2})$
* ÊãâÊôÆÊãâÊñØÊ†∏Ôºö$K(\pmb{x},\pmb{z})=exp(-\frac{\parallel \pmb{x}-\pmb{z} \parallel}{\sigma})$
* SigmoidÊ†∏Ôºö$K(\pmb{x},\pmb{z})=tanh(\beta\pmb{x}^T\pmb{z}+\theta)$

Âàô$PÊòØ‰∏Ä‰∏™m\times mÁöÑÁü©ÈòµÔºåÂÖ∂‰∏≠P_{ij}=y_iy_jK(\pmb{x_i},\pmb{x_j})$„ÄÇ

```python
# ---- Your code here ----
#ÈÄâÂèñÂ§öÈ°πÂºèÊ†∏
m = 720                          # data1.info() Ëé∑Âæó‰∏ÄÂÖ±Êúâ720‰∏™Ê†∑Êú¨
c = 0.5
mi = 5
X = np.matrix(train_data.iloc[:,0:7])
Y = np.matrix(train_data[["Class"]])
x = matrix(np.multiply(X,Y))
P = matrix((x * x.T+1)**mi)
q = matrix((-1)*np.ones((m,1)))
G1 = matrix((-1)*np.eye(m,m))
G = matrix(np.vstack((G1,-G1)))
h1 = matrix(np.zeros((m,1)))
h2 = matrix(np.ones((m,1)))*c
h = matrix(np.vstack((h1, h2)))
A = matrix(Y.T)
b = matrix(0.0)
sol=solvers.qp(P, q, G, h, A, b,kktsolver='ldl')
a = sol["x"]
# print(sol["x"])



```

<span style="color:purple">3) Ê±ÇÂá∫$b^*=y_j-\sum_{i=1}^m \alpha_i^*y_iK(\pmb{x_i},\pmb{x_j})$, ÂÖ∂‰∏≠$j$‰∏∫$\alpha^*$‰∏≠ÁöÑ‰∏Ä‰∏™Ê≠£ÂàÜÈáè$0<\alpha_j^*<C$ÁöÑ‰∏ãÊ†á„ÄÇ</span>

```python
# ---- Your code here ----
min_data = 1e-5
l = list() # list Êù•‰øùÂ≠òùõº‰∏ç‰∏∫0ÁöÑÁ¥¢Âºï
l_ = list()
for i in range(len(a)):
    if a[i]<min_data:
        a[i]=0
    if a[i]>min_data and (c - a[i])>0.001:
        l.append(i)
        l_.append(a[i])
# print(a)
print(l)
print(l_)
```

```python

j = l[0]
s = train_data[["Class"]]
y_j = s.iloc[j,0]
x_j = X[j,:].T
K = X.dot(x_j)
K += 1
K = np.array(K)
K = K**mi
sum_sample = (np.multiply(np.multiply(a,Y),K)).sum()
b_best = y_j - sum_sample
b_best
```

<span style="color:purple">4) ËØªÂÖ•ÊµãËØïÊï∞ÊçÆÈõÜ'Raisin_test.csv',Áî®ÂàÜÁ±ªÂÜ≥Á≠ñÂáΩÊï∞$f(\pmb{x})=sign(\sum_{i=1}^m \alpha_i^*y_iK(\pmb{x}_i,\pmb{x})+b^*)$ÔºàÊ≥®ÊÑèËøôÈáåÁöÑ$m,\alpha_i^*,y_i,\pmb{x}_i$ÊòØËÆ≠ÁªÉÈõÜÁöÑ, $\pmb{x}$ÊòØÊµãËØïÈõÜÁöÑÔºâËøõË°åÈ¢ÑÊµãÔºåËæìÂá∫È¢ÑÊµãÂáÜÁ°ÆÁéá„ÄÇ</span>

```python
test_data = pd.read_csv("Raisin_test.csv")
test_data
```

```python
# ---- Your code here ----
X_test =  np.matrix(test_data.iloc[:,0:7])
Y_test = np.array(test_data.iloc[:,7])
# print(Y_test)
K_test = X.dot(X_test.T)
K_test += 1
K_test = np.array(K_test)
K_test = K_test**mi
result_matrix = np.array(np.multiply(np.multiply(a,Y),K_test))
result = result_matrix.sum(axis = 0)+b_best
result_list = []
for i in result:
    result_list.append(np.sign(i))
acc = 0
for i in range(len(result_list)):
    if result_list[i] == Y_test[i]:
               acc += 1
acc = acc/180
print(" {:.2%}".format(acc))
```
